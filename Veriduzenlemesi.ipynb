```python
import cv2
import numpy as np

def GetImageFromFile(filename):
    img = cv2.imread("%s"%filename)
    return img


InputData = np.array([])
for i in range(58):
    receivedImage=0
    i=i+1
    string = 'CNNImages/%s.jpg'%i
    receivedImage = GetImageFromFile(string)
    SizedImage = cv2.resize(receivedImage,(224,224))
    print(InputData)
    InputData = np.append(InputData,SizedImage)
    print (i+1)

    
InputData = np.reshape(InputData,(-1,224,224,3))
np.save("InputData",InputData)

print(InputData)
print(InputData.shape)
```
